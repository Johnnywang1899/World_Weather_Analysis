```python
import random
import numpy as np
import pandas as pd
import requests
from citipy import citipy
from config import weather_api_key as api_key
```

```python
x = np.random.uniform(low = -90, high = 90, size = 2000)
y = np.random.uniform(low = -90, high = 90, size = 2000)
```

```python
coordinates_zip = zip(x, y)
coordinates = list(coordinates_zip)
print(coordinates)
```

```python
all_cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    if city not in all_cities:
        all_cities.append(city)
```

```python
weather_log = []
for city in all_cities:
    try:
        city_no_space = city.replace(" ", "+")
        url = f"https://api.openweathermap.org/data/2.5/weather?q={city_no_space}&appid={api_key}&units=metric"
        response = requests.get(url).json()
        country = response["sys"]["country"]
        lat = response["coord"]["lat"]
        lng = response["coord"]["lon"]
        max_temp = response["main"]["temp_max"]
       # min_temp = response["main"]["temp_min"]
        humidity = response["main"]["humidity"]
        cloudiness = response["clouds"]["all"]
        wind_speed = response["wind"]["speed"]
        current_description = response["weather"][0]["description"]
        weather_log.append({
            "City" : city,
            "Country" : country,
            "Lat" : lat,
            "Lng" : lng,
            "Max Temp" : max_temp,
            #"Min Temp" : min_temp,
            "Humidty" : humidity,
            "Cloudiness" : cloudiness,
            "Wind Speed" : wind_speed,
            "Current Description" : current_description
        })

    except KeyError:
        print("City not found. Skipping...")
        pass
```

```python
weather_df = pd.DataFrame(weather_log)
weather_df
```

```python
weather_df.to_csv("WeatherPy_Database.csv", index_label = "City_ID")
```
